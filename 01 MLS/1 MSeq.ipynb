M系列を静的に作成する関数を用意しておく

```python
'''
  M系列を静的に作成する関数
  @param 略
  @return M系列 np
'''

def createMLS(
    N = 7, #次数
    a = 3, #帰還タップ
    b = 7, #帰還タップ
    X = 1000 # M系列の個数
):
  MLS = np.empty(X, dtype = int)

  for i in range(X) :
    if i < N :
      MLS[i] = random.randrange(2)
    else :
      MLS[i] = MLS[i - a] ^ MLS[i - b]
  

  return MLS
```

## 例
次数$N = 7$,周期$T = 127$,　帰還タップ$(a, b) = (3, 7)$で乱数を$X = 1000$個作成してみる．

```python
#M系列
MlsExample = createMLS(N = 7, a = 3, b = 7, X = 1000)
print(MlsExample)
```

```python
#M系列を見る
for index in range(0, 1000 - 127, 127) :
  print("MLS[{}] ~ MLS[{}] : {}".format(
      index, index + 8, MlsExample[index: index + 7]
  ))
```

確かに一周期$2^N - 1 = 127$ごとに同じ並びになっている．

## 2 M系列中に含まれる$1$の数
作成した M 系列の関数を用いて,その M 系列の $a_i ∼ a_{i+7}(i = 0, 1, · · · , 100000)$ に含まれる $1$ の数をヒストグラムでグラフを作成する.(横軸:$1$ の数 縦軸:生起確率)

$b_i := (a_i\sim a_{i + 7}$に含まれる$1$のカウント数$)$ \\
$x_n :=(b_i == n$となる$i$のカウント数$)$ \\
$x ← \mbox{normalize}$ \\
とおき，$b$のヒストグラムを描く．

```python
b = np.array([
    np.count_nonzero(
        MlsExample[i: i + 8]
    )
    for i in range(X - 8)
])
print(b.dtype)
'''
b = np.empty(X - 8, dtype = int)
for i in range(X - 8):
  b[i] = np.count_nonzero(
      MLStask1[i: i + 8] == 1
  )

print(b)
'''
```

```python
x = np.zeros(9)
for i in range(X - 8) :
  cnt = b[i]
  x[cnt] += 1

#normalize
x /= (X - 8)
```

```python
plt.plot(x)

plt.ylim(0, 0.30)

plt.xlabel("Number of 1")
plt.ylabel("Occurrence probability")

plt.savefig("MLS_histogram.png")

plt.show()
```
